```python
import pandas as pd
```

```python
fake_euphoria = pd.read_csv('./dane-od-werki/fake_euphoria_sam.csv')
og_euphoria = pd.read_csv('./dane-od-werki/og_euphoria_sam.csv')
```

```python
fake_euphoria = fake_euphoria.iloc[:, 1:]
og_euphoria = og_euphoria.iloc[:, 1:]
```

```python
og_euphoria
```

```python
og_euphoria
```

# Original

```python
# get example to draw data
example = og_euphoria.loc[(og_euphoria['sensorId'] == 1) & (og_euphoria['cycle_step_id'] == 3)]
```

```python
example
```

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(2, 2, figsize = (12, 9))
axes = axes.flatten()

plt.title('Original Euforia')
for ax, feature in zip(axes, ['resistancy_of_sensor', 'temperature', 'pressure', 'relative_humidity']):
    ax.plot(example.index, example[feature])
    ax.set_title(feature)
```
